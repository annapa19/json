## JSON exercise


```python
#import necessary modules


import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
```

## 1. Find the 10 countries with most projects

```python
json_df=pd.read_json('data/world_bank_projects.json')
json_df[['countryname','project_name']]['countryname'].value_counts()[:10]
```

## 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
json_df['mjtheme_namecode'].value_counts()[:10]
```

We see that the 'mjtheme_namecode' column of json_df is a nested element. We then use json_normalize to create a table from this series. After the normalization, we use value_counts() on 'name' to extract the top 10 themes.

```python
themes_string=json.load((open('data/world_bank_projects.json')))
```

```python
json_normalize(themes_string,'mjtheme_namecode').name.value_counts()[:10]
```

## 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
themes=pd.DataFrame(json_normalize(themes_string,'mjtheme_namecode'))
themes
```

```python
codes=pd.DataFrame(themes['code'])
codes
```

```python
no_null=themes.drop_duplicates().replace('',np.nan).dropna(how='any')
no_null
```

To fill the entries with missing names, we create a table that shows the corresponding names for each code. We then merge this with the table 'codes'.

```python
matched=codes.merge(no_null,on='code',how='left')
matched.name.value_counts()[:10]
```
